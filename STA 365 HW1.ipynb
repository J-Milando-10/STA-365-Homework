```python
#imports and set up
from scipy.stats import beta
import numpy as np;
from scipy import stats;
import matplotlib.pyplot as plt
import random
np.random.seed(10)
```

```python
#trial one, moderately seperated thetas, decent number of runs
np.random.seed(10)
#definitions
n = 1000;

#randomly chosen, would be unknown, goal is to determine which one is highest
#(ie. give best prob of success)
theta_a = 0.30;
theta_b = 0.40;
theta_c = 0.50;

#base alpha and beta parameters, could be set to higher to skew the results of a particular
#theta to give them a headstart or handicap
alpha_a, alpha_b, alpha_c = 1,1,1;
beta_a, beta_b, beta_c = 1,1,1;

#counts the number of times a particular game (when chosen) wins or loses
win_a, win_b, win_c = 0,0,0;
loss_a, loss_b, loss_c = 0,0,0;

#randomly choose winner (1/3 chance) for first game, non informative chance
#draw random from 
for i in range(n):
    
    #the 3 betas will have their alpha and betas updated each round if they win (alpha_i) or lose (beta_i)
    test_a = beta.rvs(a = alpha_a, b = beta_a, size = 1)
    test_b = beta.rvs(a = alpha_b, b = beta_b, size = 1)
    test_c = beta.rvs(a = alpha_c, b = beta_c, size = 1)
    
    if max(test_a,test_b,test_c) == test_a: #ie. determines to play game a
        if stats.bernoulli(p= theta_a).rvs(size=1) == 1: #ie. wins game a
            win_a = win_a + 1
            alpha_a = alpha_a + 1
        else: #ie. loses game a
            loss_a = loss_a + 1
            beta_a = beta_a + 1
    elif max(test_a,test_b,test_c) == test_b: #ie. determined to play game b
        if stats.bernoulli(p= theta_b).rvs(size=1) == 1: #ie. wins game b
            win_b = win_b + 1
            alpha_b = alpha_b + 1
        else: #ie. loses game b
            loss_b = loss_b + 1
            beta_b = beta_b +1
    else: #ie. determined to play game c
        if stats.bernoulli(p= theta_c).rvs(size=1) == 1: #ie. wins game c
            win_c = win_c + 1
            alpha_c = alpha_c + 1
        else: #ie. loses game c
            loss_c = loss_c + 1
            beta_c = beta_c + 1
            
summary = [win_a + loss_a, win_b + loss_b, win_c + loss_c]
summary #sumarzes which game was chosen the most
```

```python
parameter_domain = np.linspace(0,1,1001)
posterior_a = stats.beta(a= alpha_a, b= beta_a).pdf(parameter_domain)
posterior_b = stats.beta(a= alpha_b, b= beta_b).pdf(parameter_domain)
posterior_c = stats.beta(a= alpha_c, b= beta_c).pdf(parameter_domain)


plt.plot(parameter_domain, posterior_a, label="Posterior Theta A");
plt.plot(parameter_domain, posterior_b, label="Posterior Theta B");
plt.plot(parameter_domain, posterior_c, label="Posterior Theta C"); 
plt.legend(); 
plt.title("posterior distributions after n runs, highest dist = largest theta");
```
